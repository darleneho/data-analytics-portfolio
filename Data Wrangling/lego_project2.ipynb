Your two questions are as follows:

2. In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)? Save your answer as a variable new_era in the form of an integer (e.g. 2012).

```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
df.head()
```

```python
parent_theme.head()
```

```python
# What percentage of all licensed sets ever released were 
# Star Wars themed? 
# Save your answer as a variable the_force 
# in the form of an integer (e.g. 25).

# join two differnt tables using merge syntax pandas
joined = df.merge(parent_theme, left_on = 'parent_theme', right_on = 'name')
#drop name_y and then re-make the merge table
joined.drop(columns = 'name_y', inplace=True)
joined.head()
```

```python
licensed = joined[joined['is_licensed']]
licensed.head()
```

```python
star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.head(50)
```

```python
star_wars.shape[0]
```

```python
joined[joined['set_num'].isnull()].shape
```

```python
licensed = joined[joined['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
the_force = int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```

```python
# In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)? 
# Save your answer as a variable new_era in the form of an integer (e.g. 2012).
sort_lic = licensed.sort_values('year')
licensed['count'] = 1


df_sum = sort_lic.groupby(['year', 'parent_theme']).sum().reset_index()
df_sum.head()
```

```python
max_df = df_sum.sort_values('count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df
```
