```python
# Getting all object type features
categorical_feat = []

for feat in bc.columns:
    if bc[feat].dtype == 'object':
       categorical_feat.append(feat)

# Printing all categorical features
print(f'Categorical Featuers: ')
categorical_feat
```

We split the categorical and numerical features to perform different types of data preprocessing on each type of feature.

Categorical features are variables that contain labels or categories, such as gender or marital status, and they cannot be directly used as input in most machine learning models. Therefore, we need to convert them into numerical values through a process called encoding, which allows the machine learning model to understand and use them.

```python
numerical_feat = []

for feat in bc.columns:
    if bc[feat].dtype == 'int64':
        numerical_feat.append(feat)

# Printing all numerical features
print(f'Numerical Features ')
numerical_feat
```

Numerical features, on the other hand, represent measurable quantities, such as age or tumor size, and they can be directly used as input in machine learning models. However, these features may have different scales and ranges, which can cause problems in some machine learning algorithms. Therefore, we need to scale or normalize them to bring all features to the same scale and make sure that the model treats them equally.

Splitting the features allows us to apply the appropriate data preprocessing techniques to each type of feature and improve the accuracy and performance of our machine learning model.

```python
# Applying LabelEncoder to categorical features
bc[categorical_feat] = bc[categorical_feat].apply(LabelEncoder().fit_transform)

bc.head()
```

```python
# Scale the numerical features by using RobustScaler to make the outliers less important

# Applying RobustScaler to scale orginal numerical features
bc[numerical_feat] = RobustScaler().fit_transform(bc[numerical_feat])

bc.head()
```

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, roc_auc_score

# generate a random binary classification dataset
X, y = make_classification(n_samples=800, n_classes=2, random_state=0)

# split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=0)

# create a random forest classifier and fit it to the training data
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)

# predict probabilities for the testing data
y_pred_proba = rf.predict_proba(X_test)[:,1]

# calculate the false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# calculate the area under the ROC curve
roc_auc = roc_auc_score(y_test, y_pred_proba)

# plot the ROC curve
plt.plot(fpr, tpr, label='Random Forest (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
